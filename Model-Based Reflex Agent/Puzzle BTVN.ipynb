MÔ HÌNH PEAS (8-PUZZLE AGENT)
- P (Performance): Đưa ma trận về trạng thái đích với số bước di chuyển ít nhất.
                1 2 3
                4 5 6
                8 0 7
- E (Environment): Khung ô vuông 3 x 3 chứa các số từ 1 đến 8 và một ô trống (0).
- A (Actuators): Di chuyển ô trống (Lên, Xuống, Trái, Phải).
- S (Sensors): Cảm biến hình ảnh đọc giá trị các ô số và vị trí ô trống.

PERCEPT
- Percept: Trạng thái hiện tại của bảng (State).
- Model: Lưu trữ tập các trạng thái đã khám phá (reached) để tránh lặp lại vòng lặp vô tận và lưu lộ trình từ trạng thái bắt đầu đến đích.

RULES:

Sử dụng hàng đợi FIFO để duyệt theo chiều rộng.
- Tại mỗi trạng thái, tạo ra các "con" bằng cách di chuyển ô trống.
- Nếu trạng thái con trùng với đích, trả về giải pháp.

```python
import collections
class ModelBasedReflexAgent:
    def __init__(self, start, target):
        self.initial_state = tuple(item for row in start for item in row)
        self.target = tuple(item for row in target for item in row)
        self.reached = {self.initial_state: (None, None)}
        self.frontier = collections.deque([self.initial_state])
        self.step_count = 0
    def print_board(self, state):
        """In ma trận 3x3 với khoảng trắng"""
        for i in range(0, 9, 3):
            row = state[i:i+3]
            print(f"{row[0] if row[0]!=0 else '_'}  {row[1] if row[1]!=0 else '_'}  {row[2] if row[2]!=0 else '_'}")
        print("-" * 12)
    def get_actions(self, state):
        neighbors = []
        state_list = list(state)
        idx = state_list.index(0)
        r, c = divmod(idx, 3)
        directions = [(-1, 0, "Lên"), (1, 0, "Xuống"), (0, -1, "Trái"), (0, 1, "Phải")]
        for dr, dc, label in directions:
            nr, nc = r + dr, c + dc
            if 0 <= nr < 3 and 0 <= nc < 3:
                n_idx = nr * 3 + nc
                new_s = state_list[:]
                new_s[idx], new_s[n_idx] = new_s[n_idx], new_s[idx]
                neighbors.append((tuple(new_s), label))
        return neighbors
    def run(self):
        print("Bắt đầu giải ma trận:")
        self.print_board(self.initial_state)
        found = False
        while self.frontier:
            current = self.frontier.popleft()
            self.step_count += 1
            if self.step_count <= 10:
                parent, action = self.reached[current]
                if action:
                    print(f"Bước {self.step_count}: Di chuyển ô trống {action}")
                    self.print_board(current)
            elif self.step_count == 11:
                print("... Agent tiếp tục thực hiện các bước di chuyển khác ...")
            if current == self.target:
                found = True
                break
            for child_state, action_label in self.get_actions(current):
                if child_state not in self.reached:
                    self.reached[child_state] = (current, action_label)
                    self.frontier.append(child_state)
        if found:
            print(f"\n[KẾT LUẬN]: Thành công tìm thấy đích sau {self.step_count} bước duyệt!")
        else:
            print(f"\nSau khi thử {self.step_count} trạng thái khác nhau:")
            print("[KẾT LUẬN]: Agent không thể tìm thấy đường tới đích.")
            print("Nguyên nhân: Ma trận bị nghịch thế lẻ, không thể về 7 8 0.")
start_node = [[1, 2, 3], 
              [4, 5, 6], 
              [8, 0, 7]]
target_node = [[1, 2, 3], 
               [4, 5, 6], 
               [7, 8, 0]]
agent = ModelBasedReflexAgent(start_node, target_node)
agent.run()
```
