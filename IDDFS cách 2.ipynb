```python
START_STATE = (1, 2, 3, 4, 0, 6, 7, 5, 8)
GOAL_STATE  = (1, 2, 3, 4, 5, 6, 7, 8, 0)
MOVES = {
    'U': -3,
    'D': 3,
    'L': -1,
    'R': 1
}
def get_successors(state):
    successors = []
    blank_idx = state.index(0)
    row, col = blank_idx // 3, blank_idx % 3
    for move, diff in [('U', -3), ('D', 3), ('L', -1), ('R', 1)]:
        if move == 'U' and row == 0: continue
        if move == 'D' and row == 2: continue
        if move == 'L' and col == 0: continue
        if move == 'R' and col == 2: continue
        new_idx = blank_idx + diff
        new_state = list(state)
        new_state[blank_idx], new_state[new_idx] = new_state[new_idx], new_state[blank_idx]
        successors.append(tuple(new_state))
    return successors
def state_to_str(state):
    if state == START_STATE: return "S_gốc"
    if state == GOAL_STATE: return "GOAL"
    idx = state.index(0)
    return f"N_{idx}"
def print_normal_matrix(state):
    for i in range(0, 9, 3):
        print(f"[{state[i]} {state[i+1]} {state[i+2]}]")
    print()
def print_row(node, frontier, reached, note=""):
    f_str = "[" + ", ".join([state_to_str(x[0]) for x in frontier]) + "]"
    r_str = "{" + ", ".join([state_to_str(x) for x in reached]) + "}"
    print(f"{node:<18} | {f_str:<25} | {r_str:<25} | {note}")
def depth_limited_search(state, limit):
    frontier = [(state, [state])]
    result = "cutoff"
    while frontier:
        node, path = frontier.pop()
        if node == GOAL_STATE:
            return path
        if (len(path) - 1) >= limit:
            result = "cutoff"
        else:
            for child in reversed(get_successors(node)):
                if child not in path:
                    frontier.append((child, path + [child]))
    return result
def iterative_deepening_search(start):
    depth = 0
    while True:
        result = depth_limited_search(start, depth)
        if result != "cutoff":
            return result
        depth += 1
def dls_trace(state, limit):
    frontier = [(state, 0, [state])]
    reached = set()
    print("-" * 90)
    print(f"BẮT ĐẦU CHẠY MỨC TRACE: Depth = {limit}")
    print("-" * 90)
    print(f"{'Hành động (Node)':<18} | {'F (Frontier - Stack)':<25} | {'R (Reached)':<25} | Ghi chú")
    print("-" * 90)
    print_row("(Khởi tạo)", frontier, reached)
    result = "cutoff"
    while frontier:
        node, depth, path = frontier.pop()
        reached.add(node)
        node_name = state_to_str(node)
        print_row(f"POP {node_name}", frontier, reached, f"(Nút ở depth {depth})")
        if node == GOAL_STATE:
            print_row(f"Check {node_name}", frontier, reached, "-> TRÙNG GOAL! THÀNH CÔNG.")
            return path
        print_row(f"Check {node_name}", frontier, reached, f"-> {node_name} != GOAL")
        if depth >= limit:
            print_row(f"Check Depth", frontier, reached, f"-> Depth {depth} >= {limit} (Chạm biên -> Gán cutoff)")
            result = "cutoff"
        else:
            print_row(f"Check Depth", frontier, reached, f"-> Depth {depth} < {limit} (Được phép đẻ con)")
            children = get_successors(node)
            for child in reversed(children):
                if child not in path:  
                    frontier.append((child, depth + 1, path + [child]))
            print_row("(Nạp con vào F)", frontier, reached, f"Sinh ra các con của {node_name}")
    return result
def iddfs(start):
    depth = 0
    while True:
        result = dls_trace(start, depth)
        if result != "cutoff":
            print("\n" + "="*50)
            print(f"TÌM THẤY LỜI GIẢI TẠI DEPTH = {depth}!")
            print("="*50)
            return result
        depth += 1
if __name__ == "__main__":
    print("=== THUẬT TOÁN IDS CHẠY THỰC TẾ & ĐƯỜNG ĐI LỜI GIẢI ===")
    path_solution = iterative_deepening_search(START_STATE)
    print("\n--- ĐÃ TÌM THẤY ĐƯỜNG ĐI CHI TIẾT ---")
    for idx, st in enumerate(path_solution):
        print(f"Bước {idx}:")
        print_normal_matrix(st)
    print("=" * 60 + "\n")
    print("=== MÔ PHỎNG CHI TIẾT TỪNG BƯỚC (IDDFS TRACE) ===")
    iddfs(START_STATE)
```
