Mô hình PEAS cho bài toán 8-Puzzle
- P (Performance): Số bước di chuyển ít nhất để đạt trạng thái đích, thời gian giải quyết nhanh, chi phí di chuyển thấp.
- E (Environment): Khung ô vuông 3 x  chứa các ô số từ 1 đến 8 và một ô trống (kí hiệu là 0 hoặc _).
- A (Actuators): Các hành động di chuyển ô trống: Sang trái (Left), sang phải (Right), lên trên (Up), xuống dưới (Down).
- S (Sensors): Cảm biến vị trí hiện tại của các con số trong ma trận, nhận diện trạng thái đích.

THIẾT LẬP PERCEPT VÀ RULES
Percept (Nhận thức)
- Ma trận 3 x 3 với các giá trị từ 1 -> 8 và số 0 đại diện cho ô trống.
- Ví dụ trạng thái đích (Goal):
                1 2 3
                4 5 6
                7 8 0
Rules (Quy tắc di chuyển)
Giả sử ô trống nằm tại vị trí (x, y):
- Nếu x > 0: Có thể di chuyển UP (đổi chỗ ô trống với ô phía trên).
- Nếu x < 2: Có thể di chuyển DOWN (đổi chỗ ô trống với ô phía dưới).
- Nếu y > 0: Có thể di chuyển LEFT (đổi chỗ ô trống với ô bên trái).
- Nếu y < 2: Có thể di chuyển RIGHT (đổi chỗ ô trống với ô bên phải).

```python
import random
import time
class SimpleReflexPuzzleAgent:
    def __init__(self, initial_state):
        self.grid = [list(row) for row in initial_state]
        self.goal = [[1, 2, 3], [4, 5, 6], [7, 8, 0]]
        self.m, self.n = 3, 3
    def find_blank(self):
        for i in range(self.m):
            for j in range(self.n):
                if self.grid[i][j] == 0:
                    return i, j
    def is_goal(self):
        return self.grid == self.goal
    def get_distance(self, x, y, value):
        if value == 0: return 0
        target_x, target_y = (value - 1) // 3, (value - 1) % 3
        return abs(x - target_x) + abs(y - target_y)
    def evaluate_state(self, state):
        score = 0
        for i in range(3):
            for j in range(3):
                score += self.get_distance(i, j, state[i][j])
        return score
    def show_grid(self, step_count):
        print(f"BƯỚC {step_count}:")
        for row in self.grid:
            print(" ".join(str(x) if x != 0 else "_" for x in row))
        print("-" * 20)
    def reflex_action(self):
        x, y = self.find_blank()
        possible_moves = []
        if x > 0: possible_moves.append((-1, 0, "UP"))
        if x < 2: possible_moves.append((1, 0, "DOWN"))
        if y > 0: possible_moves.append((0, -1, "LEFT"))
        if y < 2: possible_moves.append((0, 1, "RIGHT"))
        best_move = None
        min_score = float('inf')
        for dx, dy, action in possible_moves:
            nx, ny = x + dx, y + dy
            self.grid[x][y], self.grid[nx][ny] = self.grid[nx][ny], self.grid[x][y]
            score = self.evaluate_state(self.grid)
            if score < min_score:
                min_score = score
                best_move = (nx, ny, action)
            elif score == min_score:
                if random.random() > 0.5:
                    best_move = (nx, ny, action)
            self.grid[x][y], self.grid[nx][ny] = self.grid[nx][ny], self.grid[x][y]
        nx, ny, act_name = best_move
        self.grid[x][y], self.grid[nx][ny] = self.grid[nx][ny], self.grid[x][y]
        return act_name
if __name__ == "__main__":
    initial_map = [[1, 2, 3], [5, 6, 4], [8, 7, 0]]
    agent = SimpleReflexPuzzleAgent(initial_map)
    step = 0
    while not agent.is_goal() and step < 200:
        step += 1
        agent.show_grid(step)
        action = agent.reflex_action()
        print(f"==> Phản xạ: Di chuyển {action}")
        time.sleep(0.2)
    if agent.is_goal():
        agent.show_grid(step + 1)
        print(f"THÀNH CÔNG sau {step} bước!")
    else:
        print("Agent bị kẹt vào vòng lặp.")
```
