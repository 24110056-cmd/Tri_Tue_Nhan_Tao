MÔ HÌNH PEAS CỦA MÁY HÚT BỤI
- P: Làm sạch nhiều bụi nhất, ít tốn điện, ít va chạm, thời gian làm sạch nhanh
- E: Là ma trận sàn nhà mxn biểu thị các số 0 và 1, 0: sàn sạch, 1: sàn bẩn; phòng, bụi bẩn, tường, bàn ghế, vật cản
- A: Các hàm để agent hiểu và hành động (tập luật); bánh xe, chổi quét, dộng cơ hút bụi
- S: Cơ chế, hàm để agent hiểu đọc dữ liệu (cảm biến bụi, cảm biến va chạm, cảm biến khoảng cách, cảm biến vị trí)

Percept: Ma trận mxn, giá trị 0->sàn sạch, giá trị 1->sàn bẩn, 2->vị trí robot

Rules:
possible_move (x, y)
    if x>0: move.append (up)
    if x<m-1: move.append (down)
    if y>0: move.append (left)
    if y<n-1: move.append (right)
    -> return move
If state_value=0 then action=random(move)
IF state_value=1 then state-value=0 and action=random(move)

Rules:
    if state = 1 then state value=0
    if state = (0, i), i ∈ [1, n-2]
        then action = RAND(D, L, R)
    if state = (0, 0) then action = RAND(D, L)
    if state = (0, n-1) then action RAND(D, R)
    if state - (m-1, i), i ∈ [1, n-2]
        then action = RAND(U, L, R)
    if state = (m-1, 0) then action = RAND(U, L)
    if state = (m-1, n-1) then action = RAND(U, R)
    if state = (i, 0), i ∈ [1, m-2]
        then action = RAND(U, D, L)
    if state = (i, n-1), i ∈ [1, m-2]
        then action = RAND(U, D, R)
    if state = (i, j) i ∈ [1, m-2], j ∈ [1, n-2]
        then RAND(U, D, L, R)

```python
import random
import time
class RobotCleaner:
    def __init__(self, m, n):
        self.m = m
        self.n = n
        self.grid = [[random.choice([0, 1]) for _ in range(n)] for _ in range(m)]
        self.x = random.randint(0, m - 1)
        self.y = random.randint(0, n - 1)
    def is_clean(self):
        for row in self.grid:
            if 1 in row:
                return False
        return True
    def show_grid(self, step_count):
        print(f"BƯỚC {step_count}:")
        for i in range(self.m):
            row_display = []
            for j in range(self.n):
                if i == self.x and j == self.y:
                    row_display.append("R")
                else:
                    row_display.append(str(self.grid[i][j]))
            print(" ".join(row_display))
        print("-" * 20)
    def move(self):
        if self.grid[self.x][self.y] == 1:
            print(f"-> Robot phát hiện bụi tại ({self.x}, {self.y}). Đang hút...")
            self.grid[self.x][self.y] = 0
        possible_moves = []
        if self.x > 0:
            possible_moves.append("UP")
        if self.x < self.m - 1:
            possible_moves.append("DOWN")
        if self.y > 0:
            possible_moves.append("LEFT")
        if self.y < self.n - 1:
            possible_moves.append("RIGHT")
        print(f"Các hướng có thể đi: {possible_moves}")
        if possible_moves:
            action = random.choice(possible_moves)
            print(f"==> Robot chọn ngẫu nhiên bước: {action}")
            if action == "UP":
                self.x -= 1
            elif action == "DOWN":
                self.x += 1
            elif action == "LEFT":
                self.y -= 1
            elif action == "RIGHT":
                self.y += 1
            return action
        return None
if __name__ == "__main__":
    M, N = 3, 3
    robot = RobotCleaner(M, N)
    print(f"Bắt đầu robot trong lưới {M}x{N}")
    print("Ký hiệu: 1 (Bụi), 0 (Sạch), R (Robot)\n")
    step = 0
    while not robot.is_clean():
        step += 1
        robot.show_grid(step)
        action = robot.move()
        time.sleep(0.5) 
    robot.show_grid(step + 1)
    print(f"HOÀN THÀNH! Sàn đã sạch hoàn toàn sau {step} bước.")
```
