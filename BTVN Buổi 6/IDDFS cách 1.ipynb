```python
START_STATE = (1, 2, 3, 4, 0, 6, 7, 5, 8)
GOAL_STATE  = (1, 2, 3, 4, 5, 6, 7, 8, 0)
def get_successors(state):
    successors = []
    blank_idx = state.index(0)
    row, col = blank_idx // 3, blank_idx % 3
    for move, diff in [('R', 1), ('L', -1), ('U', -3), ('D', 3)]:
        if move == 'U' and row == 0: continue
        if move == 'D' and row == 2: continue
        if move == 'L' and col == 0: continue
        if move == 'R' and col == 2: continue
        new_idx = blank_idx + diff
        new_state = list(state)
        new_state[blank_idx], new_state[new_idx] = new_state[new_idx], new_state[blank_idx]
        successors.append(tuple(new_state))
    return successors
def state_to_str(state):
    if state == START_STATE: return "S_gốc"
    if state == GOAL_STATE: return "GOAL"
    return f"N_{state.index(0)}"
def print_trace_row(action_node, current_limit, status, note=""):
    node_name = state_to_str(action_node)
    print(f"{node_name:<12} | Giới hạn còn: {current_limit:<2} | Trạng thái: {status:<12} | {note}")
def recursive_dls(node, goal, limit, path):
    node_name = state_to_str(node)
    print_trace_row(node, limit, "CALL DLS", f"Kiểm tra GOAL cho {node_name}")
    if node == goal:
        print_trace_row(node, limit, "MATCH GOAL", "-> TRÙNG GOAL! THÀNH CÔNG.")
        return path
    if limit == 0:
        print_trace_row(node, limit, "CHECK LIMIT", "-> limit == 0 (Chạm biên độ sâu -> Gán cutoff)")
        return "cutoff"
    else:
        print_trace_row(node, limit, "EXPAND NODE", f"Được phép đẻ con (limit còn {limit} > 0)")
        cutoff_occurred = False
        children = get_successors(node)
        for child in children:
            if child not in path:
                result = recursive_dls(child, goal, limit - 1, path + [child])
                if result == "cutoff": cutoff_occurred = True
                elif result != "failure": return result
        if cutoff_occurred: return "cutoff"
        else: return "failure"
def print_perfect_matrices(final_path):
    print("\n" + "="*50)
    print("TÌM THẤY LỜI GIẢI TẠI DEPTH = 2 BY RECURSIVE-DLS!")
    print("="*50)
    print("\nĐÃ TÌM THẤY ĐƯỜNG ĐI:")
    for idx, state in enumerate(final_path):
        print(f"Bước {idx}:")
        print(f"[{state[0]} {state[1]} {state[2]}]")
        print(f"[{state[3]} {state[4]} {state[5]}]")
        print(f"[{state[6]} {state[7]} {state[8]}]\n")
def iterative_deepening_search(start, goal):
    depth = 0
    while True:
        print("-" * 95)
        print(f"BẮT ĐẦU CHẠY MỨC (Chạy đệ quy): Depth = {depth}")
        print("-" * 95)
        print(f"{'Nút gọi hàm':<12} | {'Mức giới hạn':<14} | {'Hành động':<12} | Ghi chú")
        print("-" * 95)
        result = recursive_dls(start, goal, depth, [start])
        if result != "cutoff" and result != "failure":
            print_perfect_matrices(result)
            return 
        depth += 1
iterative_deepening_search(START_STATE, GOAL_STATE)
```
